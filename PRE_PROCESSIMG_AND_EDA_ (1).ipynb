HISTOGRAM (to show the distribution of numerical data)

```python
#HISTOGRAM OF INCOME DISTRIBUTION
plt.figure(figsize=(8,5))
sns.histplot(df['income_annum'], kde=True)
plt.title("Income Distribution")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.show()
```

The histogram of annual income shows that applicant incomes are fairly distributed across different income ranges. The distribution does not show strong skewness or major outliers, indicating a balanced spread of income values in the dataset

```python
#HISTOGRAM OF LOAN AMOUNT DISTRIBUTION
plt.figure(figsize=(8,5))
sns.histplot(df['loan_amount'], kde=True)
plt.title("Loan Amount Distribution")
plt.xlabel("Loan Amount")
plt.ylabel("Frequency")
plt.show()
```

The loan amount distribution is positively skewed (right-skewed). Most applicants apply for small-to-medium loan amounts, while only a few applicants request very large loans. The distribution also indicates the presence of some high-value outliers

```python
#HISOGRAM OF CIBIL SCORE
plt.figure(figsize=(8,5))
sns.histplot(df['cibil_score'], kde=True)
plt.title("CIBIL Score Distribution")
plt.xlabel("CIBIL Score")
plt.ylabel("Frequency")
plt.show()
```

The CIBIL score distribution appears fairly uniform without significant skewness or extreme outliers. The dataset contains applicants with a wide range of credit scores, which is beneficial for building a robust loan approval prediction model

SCATTERPLOT  (to show the relationship between two numerical variables)

```python
#REALTIONSHIP BETWEEN INCOME_ANNUM AND LOAN_AMOUNT
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='income_annum',
    y='loan_amount',
    data=df
)
plt.title("Income vs Loan Amount")
plt.xlabel("Annual Income")
plt.ylabel("Loan Amount")
plt.show()
```

The points in the graph form an upward trend, indicating that applicants with higher incomes are generally eligible for larger loan amounts. This relationship suggests that annual income is an important factor influencing loan approval and loan size

```python
# CIBIL AND Loan Amount
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='cibil_score',
    y='loan_amount',
    data=df
)

plt.title("CIBIL Score vs Loan Amount")
plt.show()
```

The scatter plot between cibil_score and loan_amount shows that there is no strong direct relationship between CIBIL score and loan amount. The data points are widely scattered across the graph without forming a clear upward or downward trend.

This indicates that applicants with both low and high CIBIL scores may receive different loan amounts. While CIBIL score is important for loan approval decisions, it does not strongly determine the exact loan amount in this dataset.

```python
# CIBIL vs Income with Loan Status
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='cibil_score',
    y='income_annum',
    hue='loan_status',
    data=df
)

plt.title("CIBIL Score vs Income by Loan Status")

plt.show()
```

This scatter plot shows the relationship between cibil_score, income_annum, and loan_status. Different colors represent approved and rejected loan applications.

The visualization indicates that applicants with higher CIBIL scores are more likely to have approved loan status. Most approved loans are concentrated in the higher CIBIL score range, while rejected loans are more common in the lower CIBIL score range.

The plot also shows that income alone does not completely determine loan approval. Even applicants with high income may face rejection if their CIBIL score is low. Therefore, CIBIL score appears to be one of the most influential features in the loan approval prediction process.
